```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/shilnaprdisa/dataskripsi/main/datacleaning.csv'
data = pd.read_csv(url)
data
```

```python
data['date'] = pd.to_datetime(data['date'], format= "%d/%m/%Y")
```

```python
data["date"].dtype
```

```python
data["month"] = data['date'].dt.month
data["day"] = data['date'].dt.day
data.head(int(50))
```

```python
#menampilkan 10 item paling laris
plt.figure(figsize=(13,5))
sns.set_palette("muted")

sns.barplot(x = data["item"].value_counts()[:10].index,
            y = data["item"].value_counts()[:10].values)
plt.xlabel(""); plt.ylabel("")
plt.xticks(size =7, rotation = 45)
plt.title('10 Produk Paling Laris Pada Bulan Oktober', size = 20)
plt.show()
```

```python
#menampilkan data tiap hari

data['date'] = pd.to_datetime(data['date'])
data['hari'] = data['date'].dt.day_name() #menambahkan kolom hari
data_per_hari = data['hari'].value_counts()
data_per_hari = data_per_hari.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
data_perhari = data.groupby (data ['date'].dt.date) ['Transaction'].count()

plt.figure(figsize = (8,5))
sns.barplot(x=data_per_hari.index, y=data_per_hari.values, color='b')
plt.xlabel('Hari')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.xticks(size = 12, rotation = -30)
plt.title("Total Transaksi Perhari", size = 16)
```

```python
#data preparation untuk mengubah ke huruf kecil

data["item"] = data["item"]. apply(lambda item: item.lower())
```

```python
#untuk menghapus whitespace di awal dan akhir setiap string
data["item"] = data["item"]. apply(lambda item: item.strip())
```

```python
data = data[["Transaction", "item"]].copy()
data.head(10)
```

```python
#untuk menghitung jumlah barang
item_count = data.groupby(["Transaction", "item"])["item"].count().reset_index(name="Count")
item_count.head(15)
```

```python
item_count_pivot = item_count.pivot_table(index='Transaction', columns='item', values='Count', aggfunc ='sum').fillna(0)
print("Ukuran Dataset :" , item_count_pivot.shape)
item_count_pivot.head(5)
```
